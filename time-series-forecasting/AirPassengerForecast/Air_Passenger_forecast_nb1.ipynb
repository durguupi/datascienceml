```python
# Installing all the necessary packaages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("AirPassengers.csv")
```

```python
# Changing Month Coulumn to DateTimeformat
df['Month'] = pd.to_datetime(df['Month'], infer_datetime_format=True)
```

```python
df.info()
```

```python
# Setting Month as Index field

df = df.set_index(['Month'])
```

```python
df.head()
```

```python
# Plotting the data using Matplotlib Library

plt.xlabel('Date')
plt.ylabel('No of Air Passengers')
plt.plot(df);
```

```python
# Plotting against rolling mean and rolling standard deviation

rolmean = df.rolling(window=12).mean()
rolstd = df.rolling(window=12).std()

plt.plot(df, label='Original')
plt.plot(rolmean, label='Rolling Mean')
plt.plot(rolstd, label="Rolling Standard Deviation")

plt.legend(loc='best')
plt.show()
```

## Analysis from the grpah

* From the graph rolling mean has trend component
* Rolling Standard Deviation is farily constant with time
* For our time series to be stationary, we need to ensure that both rolling statistics (avg & std) remain time invariant or constant
