# Pearson corelation coefficient

`r = ∑(x - x̄)(y - ȳ) / √(∑(x - x̄)^2 ∑(y - ȳ)^2)`

where:

x and y are the two variables

x̄ and ȳ are the means of the x and y variables,

∑ (sigma) represents the sum of the values

√ (square root) represents the square root of a value

```python
# Pearson corelation coefficient using numpy

import numpy as np

# Define the variables
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Calculate the correlation coefficient
r = np.corrcoef(x, y)[0, 1]

# Print the correlation coefficient
print("Correlation coefficient:", r)
```

```python
# Pearson corelation coefficient using Pandas
import pandas as pd

# Define the variables
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a pandas DataFrame with the variables
df = pd.DataFrame({'x': x, 'y': y})

# Calculate the correlation coefficient
r = df['x'].corr(df['y'])

# Print the correlation coefficient
print("Correlation coefficient:", r)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define the variables
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a pandas DataFrame with the variables
df = pd.DataFrame({'x': x, 'y': y})

# Calculate the correlation coefficient
r = df['x'].corr(df['y'])

# Plot the variables as a scatter plot
df.plot(kind='scatter', x='x', y='y')

# Calculate the line of best fit and plot it
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

# Show the plot
plt.show()

# Print the correlation coefficient
print("Correlation coefficient:", r)
```

## Using Numpy and Matplotlib library

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the variables
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Calculate the correlation coefficient
r = np.corrcoef(x, y)[0, 1]

# Plot the variables as a scatter plot
plt.scatter(x, y)

# Calculate the line of best fit and plot it
z = np.polyfit(x, y, 2)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

# Show the plot
plt.show()

# Print the correlation coefficient
print("Correlation coefficient:", r)

```

## Pearson corelation coefficient using Pandas for negative Correlation

```python
# Pearson corelation coefficient using Pandas for negative Correlation

import pandas as pd

# Define the variables
x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]


# Create a pandas DataFrame with the variables
df = pd.DataFrame({'x': x, 'y': y})

# Calculate the correlation coefficient
r = df['x'].corr(df['y'])

# Print the correlation coefficient
print("Correlation coefficient:", r)
```

## Pearson correlation coefficient between two variables using the scipy library

```python
import numpy as np
from scipy.stats import pearsonr

# Define the variables
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Calculate the Pearson correlation coefficient
r, p = pearsonr(x, y)

# Print the Pearson correlation coefficient
print("Pearson correlation coefficient:", r)
```
